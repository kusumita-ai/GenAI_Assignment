```python
from transformers import pipeline
```

```python
ner = pipeline(
    "ner",
    model="dbmdz/bert-large-cased-finetuned-conll03-english",
    aggregation_strategy="simple"
)
```

```python
resume_text = """
John doe
Email: abc@gmail.com

Education:
B.Tech in Computer Science from PES University

Experience:
Software Intern at Apple
Worked on iOS application testing and debugging.
"""
```

```python
entities = ner(resume_text)
entities
```

```python
names = [e['word'] for e in entities if e['entity_group'] == 'PER']
organizations = [e['word'] for e in entities if e['entity_group'] == 'ORG']

print("Extracted Name(s):", names)
print("Extracted Organization(s):", organizations)
```

## Smart Resume Parser

### Project Description
This project implements a Smart Resume Parser using Named Entity Recognition (NER).
It extracts structured information such as person names and organizations from
unstructured resume text.

### Model Used
- dbmdz/bert-large-cased-finetuned-conll03-english (BERT-based NER model)

### Approach
The resume text is passed through a pre-trained NER pipeline. Detected entities
are filtered to extract Person (PER) and Organization (ORG) entities.

### Outcome
The system successfully identified candidate names, universities, and company
names, demonstrating how NLP can automate resume screening tasks.
